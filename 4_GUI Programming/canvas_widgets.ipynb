```python
# What is a canvas widget
#Drawing shapes and lines

import tkinter as tk

root = tk.Tk()
root.title("Example of Canvas")
root.geometry("400x400")

# Create canvas
canvas = tk.Canvas(
    root,
    width=400,
    height=300,
    bg="white"
)
canvas.pack()

# Draw Shapes and Lines
canvas.create_line(10, 10, 200, 200, fill="blue", width=3)
canvas.create_rectangle(50, 50, 150, 150, outline="red", width=2)
canvas.create_oval(200, 50, 350, 200, outline="green", width=2)
# Coordinates implies of bounding boxes around the shapes

# Run application
root.mainloop()
```

```python
#Mouse events on canvas

import tkinter as tk

root = tk.Tk()
root.title("Mouse Drawing")
root.geometry("400x400")

# Create canvas
canvas = tk.Canvas(
    root,
    width=400,
    height=400,
    bg="white"
)
canvas.pack()

# Draw on Mouse Drag
def draw(event):
    x, y = event.x, event.y
    canvas.create_oval(x, y, x+50, y+50, fill="yellow", outline="red")

canvas.bind("<B1-Motion>", draw)

# Run application
root.mainloop()
```

```python
#Clearing and resetting a canvas

import tkinter as tk

root = tk.Tk()
root.title("Clear Canvas")
root.geometry("500x500")

# Create canvas
canvas = tk.Canvas(
    root,
    width=400,
    height=400,
    bg="white"
)
canvas.pack()

# Draw on Mouse Drag
def draw(event):
    x, y = event.x, event.y
    canvas.create_oval(x, y, x+2, y+2, fill="yellow", outline="red")


# Clear Function
def clear_canvas():
    canvas.delete("all")

canvas.bind("<B1-Motion>", draw)

# Clear Button
clear_btn = tk.Button(
    root,
    text="Clear",
    command=clear_canvas
)
clear_btn.pack(pady=10)

# Run application
root.mainloop()
```
